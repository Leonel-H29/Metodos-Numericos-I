```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, fabs , sqrt
import random
from sympy import *


def f(x): return(x**2+x-1)
def g(x): return(1-x**2)
def dxg(x): return(-(-2*x))

def punto_fijo(x0, eps, max):
    iter = 0
    x1=x0    
    error = 2*eps+1
    print("Valor de X0", x0)
    print("Valor de X1", x1)
    print("Error: ", error)

    while(iter<=max and error>eps):
        print("-----------------------Iteracciones", iter)
        x1=g(x0)
        iter = iter+1
        error=fabs(x1-x0)
        x0=x1
        print("Valor de X0: ", x0)
        print("Valor de X1: ", x1)
        print("Error: ", error)
    if(iter>max): print("No converge en ",max, " iteracciones")
        
    else: print("Raiz= ", x1)


x = np.linspace(0,4,1000)
y1=f(x)
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title("f(x)")
plt.legend(('y'), loc='upper left')
plt.show()

x = np.linspace(0, 4, 1000)
g1 = g(x)
plt.plot(x,g1)
plt.plot(x,x)
plt.title("g1(x)")
plt.show()

x = np.linspace(0, 4, 1000)
g1_p = dxg(x)
plt.plot (x,g1_p)
#plt.plot(x,x)
plt.title("|g1'(x)|")
plt.show()
print("---------------------------------------------------------")
punto_fijo(2,10**(-6), 100)

```

```python
def Steffensen(x0, eps, max):
    iter = 0
    x1=x0+2*eps
    error = fabs(x1-x0)
    print("Valor de X0", x0)
    print("Valor de X1", x1)
    print("Error: ", error)
    while(iter<=max and error>eps):
        print("-----------------------Iteracciones", iter)
        x1=g1(x0)
        x2=g1(x1) 
        iter = iter+1
        error=fabs(x1-x0)
        x0=x0-(x1-x0)**2/(x2-2*x1+x0)
        print("Valor de X0: ", x0)
        print("Valor de X1: ", x1)
        print("Valor de X2: ", x2)
        print("Error: ", error)
    if(iter>max): print("No converge en ",max, " iteracciones")
        
    else: print("Raiz= ", x0)
        
Steffensen(1,10**(-6),100)
```

Se necesitan 4 iteracciones para llegar al valor de la raiz al cual la funcion converge.

El método de Steffensen presenta una convergencia rápida y presenta además, la ventaja adicional es permite encontrar las raíces de una función f "rápidamente" ya que tiene convergencia cuadrática.
