```python
def secante(a,b, max):
    eps = 10**(-3)
    i=0
    error=eps*2    
    XnC=0
    print("Valor de a = ",a)
    print("Valor de b = ",b)
    print("Error= ", error)

    while (error>eps ):
        XnC= a - ((b-a)/(f(b)-f(a)))*f(a)
        error = fabs(b-a)
        a =b
        b = XnC
        i = i+1
        print("----------------Iteracciones: ", i)
        print("Valor de a = ",a)
        print("Valor de b = ",b)
        print("Error= ", error)
    if(i>max): print("No Converge en ", max, i)
    else: print("Raiz = " ,XnC, "  Numero de iter= ",i)

secante(0.5, pi/2,3 )
```

La diferencia que existe entre el metodo de la secante y el metodo de regula falsi es que el de la secante toma menos tiempo de convergencia, al contrario del regula falsi en este ejemplo que necesitara un numero mas grande de interecciones para poder saber si converge pero demorara mas tiempo

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, fabs , sqrt
import random
from sympy import *


def f(x):
    return(x**2-2*x-3)
def g(x):
    return(np.square(2*x+3))

def punto_fijo(x0, eps, max):
    iter = 0
    x1=x0
    print("Valor de X0", x0)
    print("Valor de X1", x1)
    error = 2*eps
    while(iter<=max and error>eps):
        print("-----------------------Iteracciones", iter)
        x0=x1
        x1=g(x0) 
        iter = iter+1
        error=fabs(x1-x0)
        print("Valor de X0: ", x0)
        print("Valor de X1: ", x1)
        print("Error: ", error)
    if(iter>max): print("No converge en ",max, " iteracciones")
        
    else: print("Raiz= ", x1)


x = np.linspace(-4,4,100)
y1=f(x)
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title("f(x)")
plt.legend(('y'), loc='upper left')
plt.show()

x = np.linspace(0, 4, 100)
g1 = g(x)
plt.plot(x, g1)
plt.plot(x,x)
plt.title("g1(x)")
plt.show()


print("---------------------------------------------------------")
punto_fijo(-4,10**(-6), 10)
print("---------------------------------------------------------")
punto_fijo(-0.99,10**(-6), 10)
print("---------------------------------------------------------")
punto_fijo(3.5,10**(-6), 10)
print("---------------------------------------------------------")
punto_fijo(2,10**(-6), 10)
print("---------------------------------------------------------")
punto_fijo(0,10**(-6), 10)

```

```python
def g(x):
    return((x**2-3)/2)

x = np.linspace(-4.5, 8, 100)
g2 = g(x)
plt.plot(x, g2)
plt.plot(x,x)
plt.title("g2(x)")
plt.show()

print("---------------------------------------------------------")
punto_fijo(-4,10**(-3), 100)
print("---------------------------------------------------------")
punto_fijo(-0.99,10**(-3), 100)
print("---------------------------------------------------------")
punto_fijo(3.5,10**(-3), 100)
print("---------------------------------------------------------")
punto_fijo(2,10**(-3), 100)
print("---------------------------------------------------------")
punto_fijo(0,10**(-3), 100)
```
