```python
import numpy as np
import matplotlib.pyplot as mpl

def trapecio_compuesta(a, b, n, f):
    """
    
    Parametros:
    a: extremo izquierdo
    b: extremo derecho
    n: numero de intervalos
    f: funcion a integrar
    
    Retorna:
    Real: resultado de la integral definida
    """
    h = (b-a)/n
    
    sumatoria = 0
    
    for i in range(1, n):
        sumatoria += f(a + h*i)
        
    return (h/2) * (f(a) + 2 * sumatoria + f(b))

def simpson_compuesta(a, b, n, f):
    """
    
    Parametros:
    a: extremo izquierdo
    b: extremo derecho
    n: numero de intervalos
    f: funcion a integrar
    
    Retorna:
    Real: resultado de la integral definida
    """
    h = (b-a)/n
    
    sumatoria = 0
    
    for i in range(1, n):
        if (i%2 == 1):
            sumatoria += 2 * f(a + h*i)
        else:
            sumatoria += f(a + h*i)
        
    return (h/3) * (f(a) + 2 * sumatoria + f(b))

def trapecio_compuesta_sucesiva(a, b, f, epsilon, max_iter, f_metodo):
    """
    Parametros:
    f: funcion a integrar
    epsilon: error minimo
    max_iter: maxima cantidad de iteraciones
    
    Retorna:
    
    """
    i = 1
    error = 2 * epsilon
    integral_actual = f_metodo(a, b, i, f)
    
    while(i < max_iter and error >= epsilon):
        print(integral_actual)
        i += 1
        integral_anterior = integral_actual
        integral_actual = f_metodo(a, b, i, f)
        error = np.abs(integral_anterior - integral_actual)
        
    if (i >= max_iter):
        print(f"No converge en {max_iter} iteraciones")
        
    else:
        return integral_actual, i
    
def func_1(x):
    return np.sin(10*x)

xs = np.arange(0, 2, 0.01)

mpl.plot(xs, func_1(xs))
mpl.plot([0,2], [0,0])
mpl.show()


try:
    integral, iteraciones = trapecio_compuesta_sucesiva(0, 2, func_1, 10**-3, 1000, trapecio_compuesta)
except:
    pass
else:
    print(f"Valor de la integral: {integral}")
    print(f"Iteracions: {iteraciones}")
```
