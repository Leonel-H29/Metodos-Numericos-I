```python
def func1(x): return(np.sin(10*x))

def trapecio_compuesta(a, b, n, f):
    h = (b-a)/n
    
    sumatoria = 0
    
    for i in range(1, n):
        sumatoria += f(a + h*i)
        
    return (h/2) * (f(a) + 2 * sumatoria + f(b))

def trapecio_compuesta_sucesiva(a, b, f, epsilon, max_iter, f_metodo):
    
    i = 1
    error = 2 * epsilon
    integral_actual = f_metodo(a, b, i, f)
    
    while(i < max_iter and error >= epsilon):
        print("---------------------Iteracciones: ",i)
        print("--Integral actual: ",integral_actual)
        i += 1
        integral_anterior = integral_actual
        integral_actual = f_metodo(a, b, i, f)
        error = np.abs(integral_anterior - integral_actual)
        print("--Error: ", error)
        
    if (i >= max_iter):
        print(f"No converge en {max_iter} iteraciones")
        
    else:
        print("--------Valor de la integral: ", integral_actual)
        print("--------Iteracciones: ",i-1)

trapecio_compuesta_sucesiva(0, 2, func1, 10**(-6), 1000, trapecio_compuesta)
```
