```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, fabs
import random

def f(x):
    return(x**(-1)-np.tan(x))

def biseccion(a, b):
    eps=10**(-6)
    iter=0
    #p=(a+b)/2
    max=3
    print("Valor de a: ",a)
    print("Valor de b: ",b)
    print("Error soportado: ",eps)
    print("Maximo de iteracciones: ",max)
    if(f(a)*f(b))> 0:
        print("La raiz no se encuentra en el intervalo")
    else:
        while (iter <= max and fabs(a-b)>eps):
            print("-----------------Iteracciones: ", iter)
            iter=iter+1
            p= b- (f(b)*((a/b)/(f(a)-f(b))))
            if((f(a)*f(b))> 0 ): a= p
            else: b = p
            print("Valor de a: ",a)
            print("Valor de b: ",b)
            print ("Error: ",fabs(a-b))
        if(iter>max): print("No converge en ", max, iter)
        else: print("Raiz ", p)
        

plt.figure(2, (10,6),50,'c','g')
x = np.linspace(0.5,pi/2, 100)
y1=np.sin(f(x))


plt.subplot(2,1,1)
plt.plot(x,y1)
plt.show()
biseccion(0.5,pi/2)
```
