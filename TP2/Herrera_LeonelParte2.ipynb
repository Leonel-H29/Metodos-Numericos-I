```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, fabs , sqrt
import random
from sympy import *

def g(x): return(1/(x+1))
def dxg(x): return(-(-1/(x+1)**2))

def Steffensen(x0, eps, max):
    iter = 0
    x1=x0+2*eps
    error = fabs(x1-x0)
    print("Valor de X0", x0)
    print("Valor de X1", x1)
    print("Error: ", error)
    while(iter<=max and error>eps):
        print("-----------------------Iteracciones", iter)
        x1=g(x0)
        x2=g(x1) 
        iter = iter+1
        error=fabs(x1-x0)
        x0=x0-(x1-x0)**2/(x2-2*x1+x0)
        print("Valor de X0: ", x0)
        print("Valor de X1: ", x1)
        print("Valor de X2: ", x2)
        print("Error: ", error)
    if(iter>max): print("No converge en ",max, " iteracciones")
        
    else: print("Raiz= ", x0)
        
Steffensen(1,10**(-6),100)
```

Se necesitan 4 iteracciones para llegar al valor de la raiz al cual la funcion converge.

El método de Steffensen presenta una convergencia rápida y presenta además, la ventaja adicional es permite encontrar las raíces de una función f "rápidamente" ya que tiene convergencia cuadrática.



# Punto 2

```python
def f(x): return(x**3-x**2)
def dxf(x): return(3*x**2-1)
def NewtonRaphson(x0, eps, max):
    i=0
    x1= x0 + 2*eps
    error = 2*eps+1
    print("x0 = ", x0)
    print("x1 = ", x1)
    print("Error = ", error)
    print("Error maximo = ", eps)
    print("f(x0)*f(x1) ="  , f(x0)*f(x1))
    print("dxf(x0) = ", dxf(x0))
    print("")
    
    while(i<=max and error>eps):
        x1=x0-(f(x0)/dxf(x0))
        i=i+1
        error = fabs(x1-x0)
        x0=x1
        print("----------------------------Iteracciones", i)
        print("x0 = ", x0)
        print("x1 = ", x1)
        print("Error = ",error)
        print("Error maximo = ", eps)
    if(i>max): print("No converge en ", max, "iteracciones")
    else: print("Raiz: ", x1)

x = np.linspace(-1,1,1000)
y1=f(x)
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title("f(x)")
plt.legend(('y'), loc='upper left')
plt.show()

x = np.linspace(-1, 1, 1000)
f1_p = dxf(x)
plt.plot (x,f1_p)
plt.plot(x,x)
plt.title("f'(x)")
plt.show()

NewtonRaphson(np.square(1/5), (10)**(-6), 100)
```

No converge en las iteracciones ya que se aproxima a una raiz multiple (se aproxima a 0 de una manera muy lenta)
Por lo tanto utilizo el metodo de newton modificado.

```python
def dxf(x): return(6*x) #La segunda derivada de f
x = np.linspace(-1,1,1000)
y1=f(x)
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title("f(x)")
plt.legend(('y'), loc='upper left')
plt.show()

x = np.linspace(-1, 1, 1000)
f1_2p = dxf(x)
plt.plot (x,f1_2p)
plt.plot(x,x)
plt.title("f''(x)")
plt.show()
NewtonRaphson(np.square(1/5), (10)**(-6), 100)
```
