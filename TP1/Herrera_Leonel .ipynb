```python
from sympy import *
init_printing()
x,y,z = symbols('x, y, z')

Limit((exp(x)-exp(-x))/x*1/2 ,x,0)
```

```python
print("Resultado = ", Limit((exp(x)-exp(-x))/x*1/2 ,x,0).doit())
```

```python
for i in range (1,21):
    valor_x =10**(-i) #i seria el n de la funcion
    print(" n = ", i)
    print(" x = ", valor_x)
    print("Resultado = ", Limit((exp(valor_x)-exp(-valor_x))/x*1/2 ,x,valor_x).doit())
    print("---------")
    
"""" 
Cuando n crece el numero, el valor del del limite  se acerca a 1, sin ser igual a 1
""""
```

```python
"""""
f(x) = x8 − 8𝑥7 + 28𝑥6 − 56𝑥5 + 70𝑥4 − 56𝑥3 + 28𝑥2 − 8𝑥 + 1
g(x) = (𝑥 − 1)8
[0.9999, 1.001]
"""""
import numpy as np
import matplotlib.pyplot as plt
from math import pi, fabs
import random


def funcion_de_f(x):
    return(x**8-8*x**7+28*x**6-56*𝑥**5+70*𝑥**4-56*𝑥**3+28*𝑥**2-8*𝑥+1)
def funcion_de_g(x):
    return((x-1)**8)

def error_relativo(N):
    resultado= np.float32(0)
    print("Numero redondeado: " , round(N,2))
    error_abs = fabs((N-round(N,2))) #Valor absoluto entre el verdadero numero y el numero redondeado
    print("Error Absoluto: ", error_abs)
    resultado = resultado+np.float32(error_abs/N)
    return resultado;

plt.figure(2, (10,6),50,'c','g')
x = np.linspace(0.9999,1.001,100)
#y1=np.cos(funcion_de_f(x))
y1= funcion_de_f(x)*2
y2 = np.sin(funcion_de_g(x))

plt.subplot(2,1,1)
plt.plot(x,y1)
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.show()


for i in x:
    print("n = " ,i)
    print(" ")
    print("Valor de f = " ,funcion_de_f(i))
    print("Valor de g = " ,funcion_de_g(i))   
    print("Error Relativo = " ,error_relativo(i))
    print(" ")
    

    
```
