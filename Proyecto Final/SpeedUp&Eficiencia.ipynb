```python
from scipy import optimize
import random
import matplotlib.pyplot as plt

def speedUp(procesos, tiempo):
  lista = []
  """
  El primer elemento sera el 0, ya que el primer proceso
  es nuestra forma serial 
  """
  lista.append(0) 
  for i in range(1, len(procesos)):
    print(f"tiempo secuencial: {tiempo[0]}, procesos: {procesos[i]} - tiempo: {tiempo[i]}")
    """
    Calculo el Speed Up basandome de la formula:
    la division entre el tiempo serial sobre el tiempo en n procesos
    """   
    aceleracion = tiempo[0]/tiempo[i]
    lista.append(aceleracion)

  #Realizo la grafica
  print(f"Speed Up: {lista}")
  plt.figure(1, (10,6), 50, 'y', 'g')
  plt.plot(procesos,lista, marker="o", color="red")
  plt.title('SPEED UP')
  plt.legend(('Datos'), loc='upper left')
  plt.xlabel('x')
  plt.ylabel('y', fontsize=20)
  plt.grid()
  plt.show()
```

```python
def Eficiencia(procesos, tiempo):
  lista = []
  """
  El primer elemento sera el 0, ya que el primer proceso
  es nuestra forma serial 
  """
  lista.append(0)
  for i in range(1, len(procesos)):
    print(f"tiempo secuencial: {tiempo[0]}, procesos: {procesos[i]} - tiempo: {tiempo[i]}")
    """
    Calculo la eficiencia basandome de la formula:
    la division entre el tiempo serial sobre el producto
    entre los n cantidad de procesos por el tiempo en n procesos
    """ 
    eficiencia = tiempo[0]/procesos[i]*tiempo[i]
    lista.append(eficiencia)

  #Realizo la grafica
  print(f"Eficiencia: {lista}")
  plt.figure(1, (10,6), 50, 'y', 'g')
  plt.plot(procesos,lista, marker="o", color="red")
  plt.title('EFICIENCIA')
  plt.legend(('Datos'), loc='upper left')
  plt.xlabel('x')
  plt.ylabel('y', fontsize=20)
  plt.grid()
  plt.show()
```

#CON 100 ITERACIONES

```python
#Entorno virtual
procesos = [1,10,20,50]
tiempo=[0.437,0.355,0.748,3.470]

speedUp(procesos, tiempo)
Eficiencia(procesos, tiempo)
```

```python
#Entorno Local
procesos = [1,10,20,50]
tiempo=[0.255,0.388,0.521,1.354]

speedUp(procesos, tiempo)
Eficiencia(procesos, tiempo)
```
